# Data Flow diagram

```python
from graphviz import Digraph, Graph
```

```python
f = Digraph(comment='Data Flow Diagram')#, engine='neato')
#f.attr(rankdir='RL', size='5,8')


f.node('a')
f.node('b')
f.node('c')
f.node('d')

with f.subgraph(name='cluster0') as c:
    #c.attr(rankdir='RL')
    c.edges(['ab', 'ac', 'bd', 'cd'])
    
with f.subgraph(name='cluster1') as c:

    #c.attr(rankdir='RL')
    c.edge('start', 'a0')
    c.edge('start', 'b0')
    c.edge('a1', 'b3')
    c.edge('b2', 'a3')
    c.edge('a3', 'a0')
    c.edge('a3', 'end')
    c.edge('b3', 'end')

    c.node('start', shape='Mdiamond')
    c.node('end', shape='Msquare')
```
